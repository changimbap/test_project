```python
# 총 second가 들어오면
# 시간, 분, 초로 환산하는 프로그램
total = int(input("초를 입력하세요 :"))
# 만약, total이 7533초가 들어온다면
# 다음과 같이 출력
# 2시간 5분 33초
# 7533초가 -> 2시간
# 1시간 -> 3600초
hour = total // 3600
# 7533 -> 3600 x 2를 제외한 333초를 가지고 분을 계산!!
# 333 // 60
# Refactoring 
minute = total % (60 * 60) // 60
# total -> 우리가 구하고자 초!!
# 몇 달? 몇 주? 며칠? 몇 시간? 몇 분?
# 7533 -> 2시간 5분??? tatal % 3600 % 60
# 7533 -> 125분?? tatal % 60
second = total % 60
print(hour, "시간", minute, "분", second, "초")
```

```python
# 논리 연산자
print(not True)
print(not False)
print(not (10 > 5))

# 논리 곱(and)
# 1 -> True
# 0 -> False
# 1 x 1 = 1 (True and True) -> True
# 1 x 0 = 0 (True and False) -> False
# 0 x 1 = 0 (False and True) -> False
# 0 x 0 = 0 (False and False) -> False
print((10 < 5) and (3 > 1)) 
# and 앞쪽에서 이미 False이기 때문에
# and 뒷쪽은 Dead Code가 된다!! (연산 X)

# 논리 합(or)
# 1 + 1 = 2 (True or True) -> True
# 1 + 0 = 1 (True or False) -> True
# 0 + 1 = 1 (False or True) -> True
# 0 + 0 = 0 (False or False) -> False
print((10 > 5) or (3 > 1)) # True
# or 앞쪽에서 이미 True이기 때문에
# or 뒷쪽은 Dead Code가 된다!! (연산 X)
```

```python
# 추석을 기념하여
# 추석 선물세트를 구매해보자
# 1개의 가격은 10000원
# 6개 구매 -> 60000원
# 11개 이상 구매 시 10% 할인!!
# 12개 구매 -> 120,000원 -> 108,000원
count = int(input("상품의 개수를 입력 :"))
is_member = int(input("회원이라면 1, 아니라면 0 입력"))
# 60000원입니다!!
# 10% 할인하여 108000원입니다!!
price = 10000
discount = 0.1

# 1, 0, 논리값을 가지는 변수는
# is로 이름을 붙이자!!
if is_member:
    discount += 0.1

if count >= 11:
    print("10% 할인하여 {}원입니다!!".format(count * price * (1 - discount)))
else:
    print("{}원입니다!!".format(count * price))
```

# 반복문
- while (대개 반복 횟수가 정해져 있지 않을 때)
- for (대개 반복 횟수가 정해져 있을 때)

```python
# 1부터 10까지만 출력
# while 조건식:
#     실행문장
number = 1
while number <= 10:
    print(number, end=" ")
    number += 1
print()
print("프로그램이 종료되었습니다")
```

```python
# 사용자로부터 자연수를 입력 받아
# 지속적으로 누적 합하여 출력하는 프로그램
add_total = 0
# 사용자가 0을 입력하면 종료!!
# break : 가장 가까운 while, for, switch 등을 빠져나오는 키워드!!
while True:
    user_input = int(input("자연수 입력 :"))
    if user_input == 0:
        break
    add_total += user_input
    print("누적 합 :", add_total)
print("프로그램 종료")
```

```python
# 다이어트 프로그램
# 1. 현재 몸무게 입력 : 
current_weight = int(input("현재 몸무게 입력 :"))
# 2. 목표 몸무게 입력 : 
goal_weight = int(input("목표 몸무게 입력 :"))
week = 1 # 주차를 나타내는 변수

while True:
    loss_weight = int(input(f"{week}주차 감량 몸무게 :"))
    week += 1
    current_weight -= loss_weight

    if current_weight <= goal_weight:
        print(f"{current_weight}kg 달성!! 축하합니다")
        break
```

```python
# 1 ~ 100까지 수 중 하나를 맞추는 게임!!
# random : 랜덤 관련 라이브러리 
import random
answer = random.randint(1, 100)
# print(answer)

while True:
    user_digit = int(input("숫자를 입력하세요(1 ~ 100) :"))
    if user_digit > answer:
        print("입력한 숫자보다 작습니다")
    elif user_digit < answer:
        print("입력한 숫자보다 큽니다")
    else:
        print(f"{answer}!! 정답입니다!!")
        break
```
